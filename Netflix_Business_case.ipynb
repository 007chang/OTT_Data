```python
df_tv_show['ttl_ssn'] = df_tv_show['duration'].str.split(" ").str[0].astype(int,errors='ignore')
```

```python
tv_show_duration = df_tv_show.groupby('ttl_ssn')['title'].nunique().sort_values(ascending=False).reset_index()
```

```python
tv_show_duration['total_no_of_shows'] = tv_show_duration['ttl_ssn']*tv_show_duration['title']
```

```python
avg_tv_show_len = tv_show_duration['total_no_of_shows'].sum()/tv_show_duration['title'].sum()
```

```python
avg_tv_show_len
```
